# Making One Dataframe
0 False News

1 True News

```python
def make_df(df, val):
    df['text'] = df['title'] + ' ' + df['text']
    df['class'] = val
    df.drop('title', axis=1, inplace=True)
    df.drop('subject', axis=1, inplace=True)
    df.drop('date', axis=1, inplace=True)
    df = df.dropna()
    
    return df
```

```python
df_false = make_df(df_false, 0)
df_true = make_df(df_true, 1)

dataSet = pd.concat([df_false, df_true], ignore_index=True)
```

# Importing NLTK Library

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

nltk.download('stopwords')
```

```python
def clean_df(dataSet):
    n =  dataSet.shape[0]
    corpus = []
    ps = PorterStemmer()
    
    for i in range(n):
        text = re.sub('[^a-zA-Z]', ' ', dataSet['text'][i].lower())
        text = text.split()  # Spliting text into words
        text = [ps.stem(word) for word in text if not word in set(stopwords.words('english'))]  # Removing stopwords & # stemming
        text = ' '.join(text)  #Joining text
        corpus.append(text)
    
    return corpus
```

```python
#Cleaning the dataset

corpus = clean_df(dataSet)
```

```python
#Displaying the first cleaned fake article in the dataset

corpus[0]
```

# Top Words

```python
counts = {}

for article in corpus:
    for w in article.split():
        if w not in counts:
            counts[w] = 0
        counts[w] += 1
```

```python
top_words = sorted(counts, key = counts.get, reverse=True)[:10]

plt.figure(figsize=(12, 8))
for w in top_words:
    plt.bar(w, counts[w])

plt.title('Top 10 Words', size=18)
plt.xlabel('Word', size=18)
plt.ylabel('Frequency', size=18)
plt.show()
```

# Feature Extraction

```python
from sklearn.feature_extraction.text import CountVectorizer

mf = 10000
cv = CountVectorizer(max_features = mf)

X = cv.fit_transform(corpus).toarray()
y = dataSet.iloc[:, 1].values
```

# Load Data to S3

```python
import boto3
import sagemaker
import os
```

```python
# should be the name of directory you created to save your features data
data_dir = 'data'

# set prefix, a descriptive name for a directory  
prefix = 'fake-news-detection'

pd.concat([pd.DataFrame(train_x), pd.DataFrame(train_y)], axis=1) \
        .to_csv(os.path.join(data_dir, 'train.csv'), header=False, index=False)

pd.concat([pd.DataFrame(test_x), pd.DataFrame(test_y)], axis=1) \
        .to_csv(os.path.join(data_dir, 'test.csv'), header=False, index=False)
```

```python
# session and role
sagemaker_session = sagemaker.Session()
role = sagemaker.get_execution_role()

# create an S3 bucket
bucket = sagemaker_session.default_bucket()
```

```python
# upload all data to S3
input_data = sagemaker_session.upload_data(path=data_dir, bucket=bucket, key_prefix=prefix)
print(input_data)
```

# Training The Model

```python
!pygmentize train/train.py
```

```python
from sagemaker.sklearn.estimator import SKLearn

classifier = SKLearn(entry_point='train.py',
                    source_dir='train',
                    role=role,
                    train_instance_count=1,
                    train_instance_type='ml.m4.xlarge',
                    sagemaker_session=sagemaker_session)
```

```python
%%time

# Train your estimator on S3 training data
classifier.fit({'train': input_data})
```

```python
%%time

# deploy your model to create a predictor
predictor = classifier.deploy(initial_instance_count=1, instance_type='ml.m4.xlarge')
```

# Performance of the Model

```python
len(test_x)
```

```python
#Predict 2000 article at a one time

pred_y = predictor.predict(test_x[:2000, :])
pred_y = np.concatenate((pred_y, predictor.predict(test_x[2000:4000, :])))
pred_y = np.concatenate((pred_y, predictor.predict(test_x[4000:6000, :])))
pred_y = np.concatenate((pred_y, predictor.predict(test_x[6000:8000, :])))
pred_y = np.concatenate((pred_y, predictor.predict(test_x[8000:10000, :])))
pred_y = np.concatenate((pred_y, predictor.predict(test_x[10000:12000, :])))
pred_y = np.concatenate((pred_y, predictor.predict(test_x[12000:, :])))    
```

```python
from sklearn.metrics import confusion_matrix
#from sklearn.metrics import plot_confusion_matrix

cm_test = confusion_matrix(test_y, pred_y)
print(cm_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print('Accuracy:', accuracy_score(test_y, pred_y), sep='\t')
print('Precision:', precision_score(test_y, pred_y), sep='\t')
print('Recall:', recall_score(test_y, pred_y), sep='\t\t')


plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, cmap=plt.cm.Blues, fmt = 'd', annot_kws={"size": 35}).set_title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(test_y, pred_y))
```
